```python
list1 = [1, 2, 3, 4]
list2 = ['a', 'b', 'c', 'd']

import numpy as np

array1 = np.array([5, 6, 7, 8])
array2 = np.array([9, 10, 11, 12])

print(f"Type of list1: {type(list1)}")
print(f"Type of array1: {type(array1)}")

# Multiplication
product = array1 * array2

# Division
quotient = array1 / array2

# Power
power = np.power(array1, 2)

text = "Addition of Two:\n"
combined_text = text + str(array1 + array2)
print(combined_text)

import numpy as np

# Sine
sine_values = np.sin(array1)

# Logarithm (base 10)
log_values = np.log10(array2)

# Logarithm (base 2)
log2_values = np.log2(array2)

# Exponential
exp_values = np.exp(array1)

print(f"Sine values: {sine_values}")
print(f"Logarithm (base 10) values: {log_values}")
print(f"Logarithm (base 2) values: {log2_values}")
print(f"Exponential values: {exp_values}")
```
